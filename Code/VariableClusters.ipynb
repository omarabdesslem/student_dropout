```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"../Data/dropout_clean.csv")
```

```python
PlatformCluster = \
    data.columns[[17,18,19,21,23]]
print(PlatformCluster)
```

**Correlations without PlatformCluster**

```python
data = data.drop(columns = PlatformCluster)
import seaborn as sns

plt.figure(figsize=(12, 10))

sns.heatmap(
    data.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    center=0,
    square=True,
    linewidths=0.5
)

plt.title("Correlation Vector")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Rearranging the columns**

```python
columns = data.columns.tolist()
print(columns[3])
columns[3], columns[18] = columns[18], columns[3]
print(columns[3])
columns[8], columns[13], columns[19] = columns[19], columns[8], columns[13]
```

```python
data = data[columns]
import seaborn as sns

plt.figure(figsize=(12, 10))

sns.heatmap(
    data.corr(),
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    center=0,
    square=True,
    linewidths=0.5
)

plt.title("Correlation Vector")
plt.xticks(rotation=45, ha="right")
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
